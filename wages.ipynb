```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
real_wages_growth = pd.read_csv("RealWagesGrowth.csv")
labor_share_gdp = pd.read_csv("LaborShareofGDP.csv")
per_capita_wages = pd.read_csv("PerCapitaWageGrowth.csv")
tfp_growth = pd.read_csv("TFPGrowth.csv")
real_income = pd.read_csv("RealIncome.csv")
```

```python
wg_growth_pgh = real_wages_growth[real_wages_growth["variable"] == "pitt.real.growth"]
wg_growth_pgh.reset_index(inplace=True)
wg_growth_usa = real_wages_growth[real_wages_growth["variable"] == "us.real.growth"]
wg_growth_usa.reset_index(inplace=True)

fig, ax = plt.subplots(1, 1)
y_pgh = wg_growth_pgh["value"]
y_usa = wg_growth_usa["value"]
x = wg_growth_usa["year"]

ax.set_title("Pittsburgh vs. USA real wages growth")
ax.plot(x, y_pgh, color="black", label="Pittsburgh")
ax.plot(x, y_usa, color="darkkhaki", label="USA")
ax.fill_between(x, y_pgh, y_usa, where=(y_pgh > y_usa), color='black', alpha=0.3, interpolate=True)
ax.fill_between(x, y_pgh, y_usa, where=(y_pgh < y_usa), color='gold', alpha=0.3, interpolate=True)
ax.set_ylabel("Annual growth in total real wages (%)")
ax.set_xlabel("Year")
ax.legend()
```

```python
y_pgh = per_capita_wages["pgh_growth"]
y_usa = per_capita_wages["us_growth"]

x = per_capita_wages["year"]

fig, ax = plt.subplots(1, 1)

ax.set_title("Pittsburgh vs. USA per capita wages")
ax.plot(x, y_pgh, color="black", label="Pittsburgh")
ax.plot(x, y_usa, color="darkkhaki", label="USA")
ax.fill_between(x, y_pgh, y_usa, where=(y_pgh > y_usa), color='black', alpha=0.3, interpolate=True)
ax.fill_between(x, y_pgh, y_usa, where=(y_pgh < y_usa), color='gold', alpha=0.3, interpolate=True)

ax.set_ylabel("Annual growth in per capita real wages (%)")
ax.set_xlabel("Year")
ax.legend()
```

```python
lb_pgh = labor_share_gdp[labor_share_gdp["variable"] == "pitt_labor_share"]
lb_pgh.reset_index(inplace=True)
lb_usa = labor_share_gdp[labor_share_gdp["variable"] == "us_labor_share"]
lb_usa.reset_index(inplace=True)

fig, ax = plt.subplots(1, 1)
y_pgh = lb_pgh["value"]
y_usa = lb_usa["value"]
x = lb_usa["year"]

ax.set_title("Pittsburgh vs. USA labor share of GDP")
ax.plot(x, y_pgh, color="black", label="Pittsburgh")
ax.plot(x, y_usa, color="gold", label="USA")

ax.set_ylabel("Labor Share of GDP")
ax.set_xlabel("Year")
ax.legend()
```
