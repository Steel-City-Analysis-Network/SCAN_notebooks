```python
import pandas as pd
import seaborn as sns
import datetime
```

```python
df = pd.read_csv("pgh_LBSI.csv")
df.columns
```

```python
df["date_num"] = pd.to_datetime(df["Date "])
df["gdp"] = df["GDP "]
df["alpha"] = df["Alpha "]
df["L"] = df["Labor (Hours Work per year) "]
df["K"] = df["Capital Value (don\'t keep R constant "]
df["tfp"] = df["gdp"]/((df["K"]**(df["alpha"]))*(df["L"]**(1-df["alpha"])))
df["tfp"]
```

```python
sns.lineplot(data=df, x="date_num", y="gdp")
```
