```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Pittsburgh_growth_accounting.csv")
df
```

```python
df_07 = df[df["year"] >= 2007]
df_07
```

```python
df_07_growth = df_07[["year", "labor_input_growth_2007", "capital_growth_2007", "tfp_growth", "r_gdp_growth_zero_2007"]]
df_07_growth["tfp_growth"] = df["r_gdp_growth_zero_2007"] - df["capital_growth_2007"] - df["labor_input_growth_2007"]
df_07_growth = pd.melt(df_07_growth, id_vars="year", var_name="measure", value_name="Growth since 2007")
sns.set_style("darkgrid")
#g = sns.catplot(data=df_07_growth, x="year", y="Growth since 2007", hue="measure", kind="point")
#g.set_xticklabels(rotation=45)

l = sns.lineplot(data=df_07_growth, x="year", y="Growth since 2007", hue="measure")
```

```python
g = sns.catplot(data=df_07_growth, x="year", y="Growth since 2007", hue="measure", kind="bar")
g.set_xticklabels(rotation=45)
```
